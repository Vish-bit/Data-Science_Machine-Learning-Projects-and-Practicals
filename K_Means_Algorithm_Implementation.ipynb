```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#Creating Dataset

```python
from sklearn.datasets import make_blobs
#make_blobs() is the function use to create clusters of data
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

#Creating Clusters

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 4)
```

```python
model.fit(data[0])
```

```python
model.cluster_centers_
```

```python
model.labels_
#This is our cluster prediction
# data[1] - original cluster data
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c = data[1])
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c = model.labels_)
```

```python
f,(ax1, ax2) = plt.subplots(1,2, sharey = True, figsize = (10,6))
ax1.set_title('K Means predictions')
ax2.set_title('Original dataset')
ax1.scatter(data[0][:,0], data[0][:,1], c = model.labels_)
ax2.scatter(data[0][:,0], data[0][:,1], c = data[1])
```

#Elbow Method - helps to find optimal num of clusters

```python
Ks = 11  #number of centroids you want to test
mse =  np.zeros((Ks-1))  #mean squared error

for n in range(1, Ks):
  kcluster = n
  kmeans = KMeans(n_clusters = kcluster, random_state = 101, n_init = 15).fit(data[0])
  mse[n-1] = kmeans.inertia_
```

```python
plt.plot(range(1, Ks), mse)
plt.xlabel("Optimal Number of Clusters")
plt.ylabel("MSE")
plt.title("K Selection")
```

```python
# Above graph shows a elbow shape
# In graph where the elbow ends that is considered as a best value after that the model can get overfit. For above chart the best value is 4.
# High inertia -> bad
# Low inertia -> good
# Much Low intertia -> bad (Because every small group starts to form clusters)
```
